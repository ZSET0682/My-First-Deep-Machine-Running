```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
import dlib
print("ðŸŒ«ðŸ›¸")
```

```python
image_file_name = 'image2.png' # ë‚˜ì˜ ì‚¬ì§„íŒŒì¼ ì´ë¦„
my_image_path = os.getenv('HOME')+'/aiffel/camera_sticker/images/'+ image_file_name #ì´ë¯¸ì§€ ê²½ë¡œì„¤ì •
img_bgr = cv2.imread(my_image_path)    # OpenCVë¡œ ì´ë¯¸ì§€ë¥¼ ë¶ˆëŸ¬ì˜µë‹ˆë‹¤
img_show = img_bgr.copy()      # ì¶œë ¥ìš© ì´ë¯¸ì§€ë¥¼ ë”°ë¡œ ë³´ê´€í•©ë‹ˆë‹¤
plt.imshow(img_bgr)  # ì´ë¯¸ì§€ ëŒ€ê¸°
plt.show()  # ì´ë¯¸ì§€ í‘œì‹œ
```

```python
detector_hog = dlib.get_frontal_face_detector()  # face detection í• ë‹¹
print("ðŸŒ«ðŸ›¸")
```

```python
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)  # ì´ë¯¸ì§€ ëŒ€ê¸°
plt.show()  # ì´ë¯¸ì§€ í‘œì‹œ
dlib_rects = detector_hog(img_rgb, 1)   # (ì €ìž¥ëœ ì´ë¯¸ì§€(ë³€ìˆ˜), num of image pyramid)
print("ðŸŒ«ðŸ›¸")
```

```python
print(dlib_rects)   

for dlib_rect in dlib_rects:  # detector_hog(img_rgb, 1) dlib.rectangleì˜ ê°ì²´ left(), top(), right(), bottom(), height(), width()
    l = dlib_rect.left()
    t = dlib_rect.top()
    r = dlib_rect.right()
    b = dlib_rect.bottom()

    cv2.rectangle(img_show, (l,t), (r,b), (0,255,0), 2, lineType=cv2.LINE_AA)  # cv2ë¡œ ì‚¬ê°í˜• ê·¸ë¦¬ê¸°. ì™œ ì„¤ëª… ì•ˆí•´ì£¼ì§€?

img_show_rgb =  cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)  # ìˆ˜ì •í•œ ì´ë¯¸ì§€ ë³€í™˜í›„ ì €ìž¥
plt.imshow(img_show_rgb)  # ì´ë¯¸ì§€ ëŒ€ê¸°
plt.show()  # ì´ë¯¸ì§€ ëŒ€ê¸°
```

```python
model_path = os.getenv('HOME')+'/aiffel/camera_sticker/models/shape_predictor_68_face_landmarks.dat' # ëª¨ë¸ ê²½ë¡œì„¤ì •
landmark_predictor = dlib.shape_predictor(model_path)  #dlib ëžœë“œë§ˆí¬ ëª¨ë¸ í• ë‹¹
print("ðŸŒ«ðŸ›¸")
```

```python
list_landmarks = []

# ì–¼êµ´ ì˜ì—­ ë°•ìŠ¤ ë§ˆë‹¤ face landmarkë¥¼ ì°¾ì•„ëƒ…ë‹ˆë‹¤
for dlib_rect in dlib_rects:
    points = landmark_predictor(img_rgb, dlib_rect) # ì´ë¯¸ì§€ë¡œ face landmark ìœ„ì¹˜(ì¢Œí‘œ) ë§Œë“¤ê¸°
    temp = list(map(lambda p: (p.x, p.y), points.parts())) # face landmark x, y ì¢Œí‘œë¥¼ ìž„ì‹œ ì €ìž¥
    list_landmarks.append(temp)  # tempë¡œ ìž„ì‹œì €ìž¥ëœ ì¢Œí‘œë¥¼ 0ë²ˆì¸ë±ìŠ¤ì— í•œêº¼ë²ˆì— ë„£ê¸°!
    
print(len(list_landmarks[0]))
```

```python
for i in range(len(list_landmarks[0])):  # ì ì„ ëžœë“œë§ˆí¬ ê°¯ìˆ˜ë§Œí¼ ì°ì–´ì¤˜ì•¼ í•¨!
        x, y = list_landmarks[0][i]  # ì €ìž¥í•´ë‘” ì¢Œí‘œê°’ ì¨ë¨¹ê¸°.
        cv2.circle(img_show, (x, y), 5, (0, 255, 255), -1)  # ì´ë¯¸ì§€ì— ì¢Œí‘œì— ë”°ë¼ ì  ì°ê¸°

img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)  # ìˆ˜ì •í•œ ì´ë¯¸ì§€ ë³€í™˜í›„ ì €ìž¥
plt.imshow(img_show_rgb)  # ì´ë¯¸ì§€ ëŒ€ê¸°
plt.show()  # ì´ë¯¸ì§€ ì¶œë ¥
```

```python

for dlib_rect, landmark in zip(dlib_rects, list_landmarks):  # zip í•¨ìˆ˜ë¥¼ í†µí•´ ë™ì‹œì— ë‘ê°€ì§€ list ë¶ˆëŸ¬ì˜¤ê¸° 
    x = landmark[33][0]  # ì´ë¯¸ì§€ë¥¼ í‘œì‹œí•  ì¢Œí‘œ x
    y = landmark[33][1] + (landmark[51][1]-landmark[33][1]) // 2 - 5  # ì´ë¯¸ì§€ë¥¼ í‘œì‹œí•  ì¢Œí‘œ y (ì´ë¯¸ì§€ ë†’ì´ë¥¼ 33ê³¼ 51ì‚¬ì´ì— ë„£ê¸° ìœ„í•´ ì¢Œí‘œê°’ì„ ë¹¼ê³  ë‚˜ëˆ”)
    w = h = dlib_rect.width() # ì‚¬ê°í˜•ì˜ í¬ê¸°(ì¢Œí‘œê°’ ì•„ë‹˜ ì£¼ì˜)
    print ('(x,y) : (%d,%d)'%(x,y))
    print ('(w,h) : (%d,%d)'%(w,h))
```

```python
sticker_path = os.getenv('HOME')+'/aiffel/camera_sticker/images/cat-whiskers.png' # ìŠ¤í‹°ì»¤ ì´ë¯¸ì§€ ê²½ë¡œ
img_sticker = cv2.imread(sticker_path) # ìŠ¤í‹°ì»¤ ì´ë¯¸ì§€ë¥¼ ë¶ˆëŸ¬ì˜µë‹ˆë‹¤
img_sticker = cv2.resize(img_sticker, (w,h)) # ìŠ¤í‹°ì»¤ ë°•ìŠ¤ í¬ê¸°ì— ë§žì¶°ì„œ ë¦¬ì‚¬ì´ì¦ˆ
print (img_sticker.shape)
```

```python
# ì´ë¯¸ì§€ì˜ ì™¼ìª½ ìœ„ê°€ ëžœë“œë§ˆí¬ì˜ ì¢Œí‘œë¡œ ì„¤ì • ë˜ì–´ ìžˆê¸° ë•Œë¬¸ì— ì´ë¯¸ì§€ x, y ìž¬ì„¤ì •
# ì´ë¯¸ì§€ê°€ ì¤‘ê°„ì´ ì¢Œí‘œì˜ ì¤‘ê°„ê³¼ ì¼ì¹˜í•˜ë„ë¡ ìœ„ì¹˜ ì„¤ì •.
refined_x = x - w // 2
refined_y = y - h // 2
print ('(x,y) : (%d,%d)'%(refined_x, refined_y))
```

```python
# ì´ë¯¸ì§€ ìœ„ì¹˜ê°€ ìŒìˆ˜ì¼ ê²½ìš° ì´ë¯¸ì§€ë¥¼ ìžë¦„
if refined_x < 0: 
    img_sticker = img_sticker[:, -refined_x:]
    refined_x = 0
if refined_y < 0:
    img_sticker = img_sticker[-refined_y:, :]
    refined_y = 0
```

```python
sticker_area = img_bgr[refined_y:refined_y +img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]
img_bgr[refined_y:refined_y +img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]] = \
    np.where(img_sticker==0,img_sticker,sticker_area).astype(np.uint8)
plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
plt.show()
```
